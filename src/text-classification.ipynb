### **2. Classify book descriptions using Zero-shot classification**

```python
from transformers import pipeline

fiction_categories = ["Fiction", "Nonfiction"]

pipe = pipeline("zero-shot-classification",
                model="facebook/bart-large-mnli",
                device="cpu")
```

**Zero-shot Classification** is the ability to classify a piece of text (book description) into categories on which the model has never been trained directly. It does this by assessing the "fit" between the text and the category labels that have been provided.

```python
sequence = books[books['simple_categories'] == 'Fiction']['description'].reset_index(drop=True)[0]
```

```python
pipe(sequence, fiction_categories)
```

```python
max_index = np.argmax(pipe(sequence, fiction_categories)['scores'])
max_label = pipe(sequence, fiction_categories)['labels'][max_index]
max_label
```

```python
def generate_predictions(sequence, categories):
  predictions = pipe(sequence, categories)
  max_index = np.argmax(pipe(sequence, categories)['scores'])
  max_label = pipe(sequence, categories)['labels'][max_index]
  return max_label
```

### **3. Checking classifier accuracy**

```python
from tqdm import tqdm

actual_categories = []
predicted_categories = []

for i in tqdm(range(0, 300)):
  sequence = books[books['simple_categories'] == 'Fiction']['description'].reset_index(drop=True)[i]
  predicted_categories += [generate_predictions(sequence, fiction_categories)]
  actual_categories += ['Fiction']
```

In this part, we will evaluate the performance of Zero-shot classification model on the labeled data `simple_categories`. Taking 300 descriptions of **Fiction** and 300 descriptions of **Nonfiction**, after that we will check the prediction of the model. This helps to determine whether the model is reliable before applying it to books without a label.

```python
df_predict_fiction = pd.DataFrame({
    'actual_categories': actual_categories,
    'predicted_categories': predicted_categories
})
```

```python
for i in tqdm(range(0, 300)):
  sequence = books[books['simple_categories'] == 'Nonfiction']['description'].reset_index(drop=True)[i]
  predicted_categories += [generate_predictions(sequence, fiction_categories)]
  actual_categories += ['Nonfiction']
```

```python
df_predict_nonfiction = pd.DataFrame({
    'actual_categories': actual_categories,
    'predicted_categories': predicted_categories
})
```

```python
df_combined_predictions = pd.read_csv("df_combined_predictions.csv")
df_combined_predictions.head()
```

```python
df_combined_predictions['correct_predictions'] = (
    np.where(df_combined_predictions['predicted_categories'] == df_combined_predictions['actual_categories'], 1, 0)
)
```

```python
df_combined_predictions.head()
```

```python
print("The accuracy of the testing:", df_combined_predictions['correct_predictions'].sum()/ len(df_combined_predictions))
```

The result shows that the zero-shot classification model is quite reliable to apply to books that do not have labels.

### **4. Labeling the books that are missing categories**

```python
# Take the identities and descriptions of books that are missing labels
missing_categories = books[books['simple_categories'].isna()][['isbn13', 'description']].reset_index(drop=True)
```

```python
missing_categories.head()
```

```python
from tqdm import tqdm

isbns = []
predicted_categories = []

for i in tqdm(range(0, len(missing_categories))):
  sequence = missing_categories['description'][i]
  predicted_categories += [generate_predictions(sequence, fiction_categories)]
  isbns += [missing_categories['isbn13'][i]]
```

```python
df_missing_predicted = pd.DataFrame({"isbn13": isbns, "predicted_categories": predicted_categories})
```

```python
df_missing_predicted.head()
```

```python
books = pd.merge(books, df_missing_predicted, on="isbn13", how="left")
books["simple_categories"] = np.where(books["simple_categories"].isna(), books["predicted_categories"], books["simple_categories"])
books = books.drop(columns = ["predicted_categories"])
```
